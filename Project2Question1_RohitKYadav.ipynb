## Question 1: Exploring World Happiness

According to Wikepedia, the International Day of Happiness is celebrated worldwide every March 20, and was conceptualized and founded by 
philanthropist, activist, 
statesman, and prominent United Nations special advisor Jayme Illien to inspire, mobilize, and advance the global happiness movement.

In 2011, Illien brought the idea and concept of creating a new global day of awareness, the International 
Day of Happiness, to senior United Nations Officials.

The first World Happiness Report was published in April, 2012, in support of the UN High Level Meeting on happiness and well-being. 
Increasingly, happiness is considered to be the proper measure of social progress and 
the goal of public policy. In June 2016 the OECD committed itself “to redefine the growth narrative to put people’s well-being at 
the center of governments’ efforts”. In February 2017, the United Arab Emirates held a full-day World Happiness meeting, as
part of the World Government Summit.

# K-Means Clustering on Happiness 2015 Dataset

Doing K-Means analysis with 6 clusters.

```r
result_15 <- kmeans(happy_15, 6)
result_15
```

Result shows the 'means' of the generated 3 clusters for all the features in the dataset. 

```r
table(happy_15$Family, result_15$cluster)
```

# Plotting clusters based on kmeans analysis

```r
plot(happy_15[c("Economy..GDP.per.Capita.","Dystopia.Residual")], col = result_15$cluster)
```

# Plotting clusters based on actual data

```r
plot(happy_15[c("Economy..GDP.per.Capita.","Dystopia.Residual")], col = original_happy_15$Happiness.Score)
```

# Gaussian Mixture on Happiness-2015 dataset.

```r
citation("mclust")
library(mclust)
```

```r
cluster_happy_15 <- Mclust(happy_15)
```

```r
names(cluster_happy_15)
```

```r
cluster_happy_15$classification
```

We can see the names of the countries that belongs to each of these clusters. We have identified some of the countries belonging to these 4 generated clusters. It is represented in the table below.

```r
countries_cluster <- data.frame(Cluster1 = c("Switzerland", "Iceland", "Denmark", "Norway", "Canada"), 
                 Cluster2 = c("United States", "Luxembourg", "United Arab Emirates", "Oman", "Singapore"), 
                 Cluster3 = c("Israel", "Costa Rica", "Mexico", "Brazil", "Venezuela"), 
                 Cluster4 = c("Togo", "Burundi", "Benin", "Afghanistan", "Burkina Faso"))
countries_cluster
```

According to the results we can conclude that cluster_1 countries are the top countries in happiness index whereas cluster_4 countries are the last countries in happiness index. So It proofs that the clustering method is working for this given dataset.

```r
summary(cluster_happy_15)
```

# Seriation analysis on 2015 dataset

```r
library(seriation)
x <- as.matrix(happy_15)
d <- dist(x)
o <- seriate(d)
pimage(d, main='original')
pimage(d, o, main='reordered')
```

```r
get_order(o)
```

After using seriation analysis, we have found that the order is Mexico (14) followed by Central African Republic (148) while Norway (4) is in 29th place.

```r
hmap(x)
```
