## Question 1: Exploring World Happiness

```r
original_happy_15 <- read.csv("https://raw.githubusercontent.com/sauchilee/Stat517/master/Data/World_Happiness_2015.csv")
```

```r
names(original_happy_15)
```

```r
happy_15 <- original_happy_15[5:10]
happy_15
```

# K-Means Clustering on Happiness 2015 Dataset

Doing K-Means analysis with 5 clusters.

```r
result_15 <- kmeans(happy_15, 5)
result_15
```

```r
table(happy_15$Family, result_15$cluster)
```

# Plotting clusters based on kmeans analysis

```r
plot(happy_15[c("Economy..GDP.per.Capita.","Health..Life.Expectancy.")], col = result_15$cluster)
```

# Plotting clusters based on actual data

```r
plot(happy_15[c("Economy..GDP.per.Capita.","Trust..Government.Corruption.")], col = original_happy_15$Happiness.Score)
```

# Seriation analysis on 2015 dataset

```r
library(seriation)
x <- as.matrix(happy_15)
d <- dist(x)
o <- seriate(d)
pimage(d, main='original')
pimage(d, o, main='reordered')
```

```r
get_order(o)
```

we have found that the order is Mexico (14) followed by Central African Republic (148) while Norway (4) is in 29th place.

```r
hmap(x)
```

# Seriation analysis on 2016 dataset

```r
x <- as.matrix(happy_16)
d <- dist(x)
o <- seriate(d)
pimage(d, main='original')
pimage(d, o, main='reordered')
```

```r
get_order(o)
```

using seriation analysis, we found that the order is Denmark (1) followed by Switzerland (2) whereas Norway (4) is in 4th place.

```r
hmap(x)
```

# Seriation analysis on 2017 dataset
